### Patterns 

##### print(&) in stright line with gap

```python
for n in range(1,9):
    for i in range(2):
        print('&')
    print( )
```

##### Print(*) in stright line

```python
for n in range(1,8):
    for i in range(1):
        print('*',end=' ')
        i+=1
    print( )
```

##### Right triangle with pattern(&,#)

```python
 for i in range(1,10):
     if i%2==0:
         print('# '*i)
     else:
         print('& '*i)
```

```python
for i in range(1,10):
    if i%4==0:
        print('@' *i)
    else:
        print('*' *i)
```

```python
for r in range(1,12):
    if r%3==0:
        print('$ '*r)
    else:
        print('% '*r)
```

```python
for i in range(1,8):
    print("i "*i)
```

#### Left triangle pattern 

```python
n=7
for i in range(n):
    for j in range(n -i):
        print(" ",end=" ")
    for j in range(i +1):
        print("*",end=" ")
    print( )
```

```python
D=5
for i in range(D):
    for k in range(D -i):
        print(" ",end=" ")
    for k in range(i +1):
        print("&",end=" ")
    print( )
```

```python
E=9
for i in range(E):
    for k in range(E -i):
        print(" ",end=" ")
    for k in range(i +1):
        print(i +1,end=" " )
    print( )
```

##### Square pattern

```python
a=6
for i in range(a):
    for j in range(a):
        print("*",end=" ")
    print( )
```

#### Reverse pattern

```python
n=6
for i in range(n,0,-1):
    m=i
    for j in range(0,i):
        print(m,end=" ")
    print( )
```

```python
G=10
for i in range(G,0,-1):
    H=G
    for K in range(0,-1):
        print(G,end=" ")
    print( )
```
